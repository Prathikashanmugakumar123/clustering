```python
import numpy as np
from sklearn_extra.cluster import KMedoids

X = np.array([10, 12, 20, 22, 50, 52]).reshape(-1, 1)

model = KMedoids(n_clusters=2, random_state=0, metric="manhattan")

labels = model.fit_predict(X)

print("Data:", X.flatten())
print("Medoids:", model.cluster_centers_.flatten())


for i in range(len(X)):
    print(X[i][0], "-> Cluster", labels[i])
```
