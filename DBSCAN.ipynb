```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

X = np.array([2, 3, 8, 10]).reshape(-1, 1)

model = DBSCAN(eps=2, min_samples=2)
labels = model.fit_predict(X)

for i in range(len(X)):
    print("Shop", chr(65+i), "-> Cluster", labels[i])

plt.scatter(X, [0,0,0,0])
plt.xlabel("Distance (km)")
plt.title("DBSCAN Shops")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors

X = np.array([2, 3, 8, 10]).reshape(-1, 1)

k = 2   
nn = NearestNeighbors(n_neighbors=k)
nn.fit(X)

distances, _ = nn.kneighbors(X)
eps_auto = distances[:, k-1].max()   


model = DBSCAN(eps=eps_auto, min_samples=2)
labels = model.fit_predict(X)

print("Automatic epsilon:", eps_auto)
for i in range(len(X)):
    print("Shop", chr(65+i), "-> Cluster", labels[i])

plt.scatter(X, np.zeros(len(X)))
plt.xlabel("Distance (km)")
plt.title("DBSCAN Shops (Automatic Epsilon)")
plt.show()
```
